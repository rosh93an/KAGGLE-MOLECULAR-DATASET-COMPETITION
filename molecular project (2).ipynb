 RDKit Feature Engineering ---

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
```

```python
def generate_rdkit_features(smiles_list):
    features = []
    for smi in smiles_list:
        mol = Chem.MolFromSmiles(smi)
        if mol is None:
            features.append([np.nan] * 5)
            continue
        f = [
            Descriptors.MolWt(mol),                  # Molecular Weight
            Descriptors.NumHDonors(mol),             # H-bond Donors
            Descriptors.NumHAcceptors(mol),          # H-bond Acceptors
            Descriptors.TPSA(mol),                   # Polar Surface Area
            Descriptors.NumRotatableBonds(mol)       # Flexibility
        ]
        features.append(f)
    return np.array(features)

rdkit_feats = generate_rdkit_features(smiles)
rdkit_feat_df = pd.DataFrame(rdkit_feats, columns=["MolWt", "HDonors", "HAcceptors", "TPSA", "RotBonds"])
```

```python
X_augmented = pd.concat([X.reset_index(drop=True), rdkit_feat_df], axis=1)
```

```python
X_train_aug, X_val_aug, y_train, y_val = train_test_split(X_augmented, Y, test_size=0.2, random_state=42)
```

```python
selector = SelectKBest(score_func=f_regression, k=25)
X_train_selected = selector.fit_transform(X_train_aug, y_train)
X_val_selected = selector.transform(X_val_aug)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_log_error
```

```python
param_dist = {
    'n_estimators': randint(350, 450),          # center around 400
    'max_depth': randint(4, 7),                 # center around 5
    'learning_rate': uniform(0.3, 0.6),         # 0.3 to 0.6
    'subsample': uniform(0.8, 0.2),             # 0.8 to 1.0
    'reg_alpha': uniform(4, 3),                 # 4 to 6
    'reg_lambda': uniform(0.5, 1.0),            # 0.5 to 1.5
    'colsample_bytree': uniform(0.7, 0.3)       # 0.7 to 1.0
}
```

```python
xgb_model2 = XGBRegressor(random_state=40)
```

```python
grid_search = GridSearchCV(
    estimator=xgb_model2,
    param_grid=param_grid,
    scoring='neg_mean_squared_log_error',
    cv=5,
    verbose=1,
    n_jobs=-1
)
```

```python
grid_search.fit(X_train_selected, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
best_model.fit(X_train_selected, y_train)
```

```python
y_pred = best_model.predict(X_val_selected)
```
